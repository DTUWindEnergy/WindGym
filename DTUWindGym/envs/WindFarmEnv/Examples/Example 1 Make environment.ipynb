```python
#Load dependencies
import gymnasium as gym
from DTUWindGym.envs import WindFarmEnv
from py_wake.examples.data.hornsrev1 import V80
```

```python
#The environment can be created either with the gym.make() function or by directly calling the WindFarmEnv class

#We can create a new environment with the following code
env1 = gym.make('DTUWindGym/WindFarmEnv-v0', turbine=V80(), 
                 yaml_path="EnvConfigs/2turb.yaml", render_mode="human")

#Or we can create a new environment with the following code
env2 = WindFarmEnv(turbine=V80(), 
                 yaml_path="EnvConfigs/2turb.yaml")

#Note that the render_mode="human" maked the environment render the wind farm in a human readable way.
#That is super slow
```

```python
obs = env1.reset() 
for _ in range(20):                     # Repeat for X steps
    a = env1.action_space.sample()      # Choose an action randomly
    obs, reward, terminated, truncated, info = env1.step(a)  # Perform the action
env1.close()
```

```python
#For the render function to work, we need to set the render_mode="human"
obs = env2.reset() 
for _ in range(20):                     # Repeat for 1000 steps
    a = env2.action_space.sample()      # Choose an action randomly
    obs, reward, terminated, truncated, info = env2.step(a)  # Perform the action
    env2.render()                       # Render the environment; Note that 
env2.close()
```
