```python
#Load dependencies
import gymnasium as gym
from WindGym import WindFarmEnv
from py_wake.examples.data.hornsrev1 import V80
```

```python
# For more information on the environment, use the help function
help(WindFarmEnv)
```

```python
# The environment can be created by directly calling the WindFarmEnv class

env = WindFarmEnv(turbine=V80(),            # The turbine type. Must be a py_wake Turbine object
                 n_passthrough = 10,        # The ammount of times the wind flows over the whole farm, before a new episode starts
                 turbtype = "Random",       # Just use Random turbulence
                 yaml_path="EnvConfigs/Env1.yaml") #Most of the details are specified from the environment yaml file
```

```python
obs = env.reset() 
for _ in range(20):                     # Repeat for X steps
    a = env.action_space.sample()       # Choose an action randomly
    obs, reward, terminated, truncated, info = env.step(a)  # Perform the action
env.close()
```
